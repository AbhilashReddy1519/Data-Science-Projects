# Creating DataFrame in Pandas

## Creating DataFrame from an Array

### option 1

```python
# creating an array
data = np.array([[1,4],[2,5],[3,6]])
```

```python
# creating dataframe 
df = pd.DataFrame(data)
# showing dataframe with assigned values for index and columns
df
```

```python
# here index and columns are variables needs to in same name
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'], columns=['col1', 'col2'])
```

```python
# showing dataframe
df
```

### option 2

```python
#creating array with list shape
data = [[1, 4], [2, 5], [3, 6]]
```

```python
# creating a dataframe
df = pd.DataFrame(data, index=['row1', 'row2', 'row3'],
                 columns=['col1', 'col2'])
```

```python
# showing the datafrane
df
```

## Creating a dataframe from Dictionary

```python
# lists used from value for dictionary keys
states = ["Andhra Pradesh", "Karnataka", "Telangana","Tamilnadu"]
population = [3904398403,2490240249,209204024,209240220] # random population values
```

```python
# creating dictionary and storing list in it
dict_states = {"States": states, "Population": population}
```

```python
# creating dataframe
df_sp = pd.DataFrame(dict_states)
# df_sp = pd,DataFrame.from_dict(dict_states)
```

```python
# showing the datafrane
df_sp
```
